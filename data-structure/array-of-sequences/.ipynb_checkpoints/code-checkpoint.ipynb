````python
from IPython.display import Markdown, display
from collections.abc import MutableSequence, Sequence, Container, Iterable, Sized

issubclass(MutableSequence, Sequence), issubclass(Sequence, Container)
````

````python
markdown = '''
```mermaid
classDiagram
    class Container {
        <<abstract>>
        +__contains__(item)*
    }
    
    class Iterable {
        <<abstract>>
        +__iter__()*
    }
    
    class Sized {
        <<abstract>>
        +__len__()*
    }
    
    class Collection {
        <<abstract>>
        +__contains__
        +__iter__
        +__len__
    }
    
    class Sequence {
        <<abstract>>
        +__getitem__*
        +__contains__
        +__iter__
        +__len__
        +__reversed__
        +count()
        +index()
    }
    
    class MutableSequence {
        <<abstract>>
        +__setitem__*
        +__delitem__*
        +insert()*
        +append()
        +clear()
        +extend()
        +pop()
        +remove()
        +reverse()
    }
    
    class Reversible {
        <<abstract>>
        +__reversed__*
    }
    
    Container <|-- Collection
    Iterable <|-- Collection
    Sized <|-- Collection
    Collection <|-- Sequence
    Reversible <|-- Sequence
    Sequence <|-- MutableSequence
    
    note for Collection "Python 3.6+ 引入\n组合 Container, Iterable, Sized"
    note for Sequence "有序集合\n支持下标访问"
    note for MutableSequence "可变序列\n支持修改操作"
```
'''

display(Markdown(markdown))
````
