```python
from collections import namedtuple

Model = namedtuple('Model', 'vendor name released')

class OpenRouter:
    def __init__(self):
        models = {
            'openai':[
                ('gpt-5.2-pro', 'Dec 10, 2025'),
                ('gpt-5.1', 'Nov 13, 2025'),
                ('gpt-5-mini', 'Aug 7, 2025'),
            ],
            'anthropic':[
                ('claude-sonnet-4.6', 'Feb 17, 2026'),
                ('claude-opus-4.6', 'Feb 4, 2026'),
            ],
            'google':[
                ('gemini-3.1-pro-preview', 'Feb 19, 2026'),
            ]
        }
        self._models = [Model(vendor, name, released) for vendor in models.keys() for name, released in models[vendor]]

    def __len__(self):
        return len(self._models)

    def __getitem__(self, pos):
        return self._models[pos]


router = OpenRouter()

len(router)
```

```python
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __repr__(self):
        return f'Money({self.amount!r})'

    def __str__(self):
        return f'{self.amount!s}CNY'
    
    def __add__(self, other):
        return Money(self.amount + other.amount)

    def __mul__(self, scaler):
        return Money(self.amount * scaler)

    def __rmul__(self, scaler):
        return self * scaler
    
    def __bool__(self):
        return self.amount != 0

m1 = Money(12)
m2 = Money(21)

m1 + m2
```
